```python
import pandas as pd
import numpy as np

from sklearn.ensemble import AdaBoostRegressor,AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor

#Train ve Test Setlerinin ayrılması
from sklearn.model_selection import train_test_split

#Değerlendirme Metrikleri
from sklearn.metrics import accuracy_score,classification_report ,mean_absolute_error,mean_squared_error,r2_score

#Cross-validation ve en iyi parametrelerin seçimi
from sklearn.model_selection import GridSearchCV

# Gereksiz uyarıların görüntülenmemesi için kullanıyoruz
import warnings
warnings.filterwarnings('ignore')
```

# Adaboost Classifier İle Sınıflandırma

```python
df = pd.read_csv('./Data/breast-cancer-wisconsin.csv')
df
```

```python
df.info()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
ada_clf = AdaBoostClassifier(estimator=DecisionTreeClassifier())
ada_clf_param_grid = {
    'n_estimators':[20,50,100,200,300],
    'learning_rate':[0.001, 0.01,0.1,1,3,5,10]        
}
```

```python
ada_clf_cv = GridSearchCV(ada_clf,param_grid=ada_clf_param_grid,cv=5,verbose=3)
ada_clf_cv.fit(X_train,y_train)
```

```python
ada_clf_cv.best_params_
```

```python
print('='*100)
print('Test Hatası')
print('='*100)
y_pred = ada_clf_cv.predict(X_test)
print(classification_report(y_test,y_pred))
print('='*100)
print('Train Hatası (Ezber Kontrolü)')
print('='*100)
y_pred_tr= ada_clf_cv.predict(X_train)
print(classification_report(y_train,y_pred_tr))
```

# AdaBoost İle Regresyon

```python
df2 = pd.read_csv('./Data/insurance.csv')
df2
```

```python
df2.info()
```

```python
df2_encoded = pd.get_dummies(df2,columns=['sex','smoker','region'],dtype=np.int64)
X2 = df2_encoded.drop(columns=['charges'])
y2 = df2_encoded['charges']
X2_train,X2_test,y2_train,y2_test = train_test_split(X2,y2,train_size=0.8,random_state=42)
```

```python
ada_reg = AdaBoostRegressor(DecisionTreeRegressor())
ada_reg_grid_params = {
    'n_estimators':[20,50,100,200,300],
    'learning_rate':[0.001, 0.01,0.1,1,3,5,10] ,
    'loss': ['linear','square','exponential']
}
```

```python
ada_reg_cv = GridSearchCV(ada_reg,param_grid=ada_reg_grid_params,cv=3,verbose=3)
ada_reg_cv.fit(X2_train,y2_train)
```

```python
ada_reg_cv.best_params_
```

```python
print('='*100)
print('Test Hatası')
print('='*100)

y2_pred = ada_reg_cv.predict(X2_test)
print("MAE:",mean_absolute_error(y2_test,y2_pred)) 
print("MSE:",mean_squared_error(y2_test,y2_pred))
print("RMSE:",np.sqrt(mean_squared_error(y2_test,y2_pred)))
print("R2 Score:",r2_score(y2_test,y2_pred))

print('='*100)
print('Train Hatası (Ezber Kontrolü)')
print('='*100)

y2_pred_train = ada_reg_cv.predict(X2_train)
print("MAE:",mean_absolute_error(y2_train,y2_pred_train)) 
print("MSE:",mean_squared_error(y2_train,y2_pred_train))
print("RMSE:",np.sqrt(mean_squared_error(y2_train,y2_pred_train)))
print("R2 Score:",r2_score(y2_train,y2_pred_train))
```
