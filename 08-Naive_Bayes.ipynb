```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Model
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import accuracy_score,classification_report 

#Cross-validation ve en iyi parametrelerin seçimi
from sklearn.model_selection import GridSearchCV,train_test_split

#Kategorik Değişken Dönüşümü
from sklearn.preprocessing import LabelEncoder

# Gereksiz uyarıların görüntülenmemesi için kullanıyoruz
import warnings
warnings.filterwarnings('ignore')
```

# Naive Bayes ile Sınıflandırma

```python
df = pd.read_csv('./Data/breast-cancer-wisconsin.csv')
df
```

```python
df.info()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
# 10 ile 1*10^-9 arasında 100 sayı oluşturur
np.logspace(1,-9, num=100)
```

```python
nb_clf = GaussianNB()
NB_clf_params = {'var_smoothing': np.logspace(0,-9, num=100)}
```

```python
nb_clf_cv = GridSearchCV(nb_clf,param_grid=NB_clf_params,cv=5,verbose=3)
nb_clf_cv.fit(X_train,y_train)
```

```python
nb_clf_cv.best_params_
```

```python
y_pred = nb_clf_cv.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
y_pred_tr = nb_clf_cv.predict(X_train)
print(classification_report(y_train,y_pred_tr))
```

```python
# Parametrelerimizi bulduktan sonra modelimizi bu şekilde de eğitebiliriz
nb = GaussianNB(var_smoothing=2.310129700083158e-09)
nb.fit(X_train,y_train)
```

```python
y_pred2 = nb.predict(X_test)
print(classification_report(y_test,y_pred2))
```
