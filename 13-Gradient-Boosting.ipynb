```python
import pandas as pd
import numpy as np

from sklearn.ensemble import GradientBoostingClassifier,GradientBoostingRegressor
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor

#Train ve Test Setlerinin ayrılması
from sklearn.model_selection import train_test_split

#Değerlendirme Metrikleri
from sklearn.metrics import accuracy_score,classification_report ,mean_absolute_error,mean_squared_error,r2_score

#Cross-validation ve en iyi parametrelerin seçimi
from sklearn.model_selection import GridSearchCV

# Gereksiz uyarıların görüntülenmemesi için kullanıyoruz
import warnings
warnings.filterwarnings('ignore')
```

# GradientBoosting Classifier İle Sınıflandırma

```python
df = pd.read_csv('./Data/breast-cancer-wisconsin.csv')
df
```

```python
df.info()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
gbm_clf = GradientBoostingClassifier()
gbm_clf_param_grid = {
    'n_estimators':[20,50,100,200,300],
    'learning_rate':[0.001, 0.01,0.1,1,5],
    'max_depth' : [5,7,9],
    'min_samples_split': [2,3]
}
```

```python
gbm_clf_cv = GridSearchCV(gbm_clf,param_grid=gbm_clf_param_grid,cv=3,verbose=3)
gbm_clf_cv.fit(X_train,y_train)
```

```python
gbm_clf_cv.best_params_
```

```python
print('='*100)
print('Test Hatası')
print('='*100)
y_pred = gbm_clf_cv.predict(X_test)
print(classification_report(y_test,y_pred))
print('='*100)
print('Train Hatası (Ezber Kontrolü)')
print('='*100)
y_pred_tr= gbm_clf_cv.predict(X_train)
print(classification_report(y_train,y_pred_tr))
```

# GBM ile Regresyon

```python
df2 = pd.read_csv('./Data/insurance.csv')
df2
```

```python
df2.info()
```

```python
df2_encoded = pd.get_dummies(df2,columns=['sex','smoker','region'],dtype=np.int64)
X2 = df2_encoded.drop(columns=['charges'])
y2 = df2_encoded['charges']
X2_train,X2_test,y2_train,y2_test = train_test_split(X2,y2,train_size=0.8,random_state=42)
```

```python
gbm_reg = GradientBoostingRegressor()
gbm_reg_param_grid = {
    'n_estimators':[20,50,100,200,300],
    'learning_rate':[0.001, 0.01,0.1,1,5],
    'max_depth' : [5,7,9],
    'min_samples_split': [2,3]
}
```

```python
gbm_reg_cv = GridSearchCV(gbm_reg,param_grid=gbm_reg_param_grid,cv=3,verbose=3)
gbm_reg_cv.fit(X_train,y_train)
```

```python
gbm_reg_cv.best_params_
```

```python
print('='*100)
print('Test Hatası')
print('='*100)

y2_pred = gbm_reg_cv.predict(X2_test)
print("MAE:",mean_absolute_error(y2_test,y2_pred)) 
print("MSE:",mean_squared_error(y2_test,y2_pred))
print("RMSE:",np.sqrt(mean_squared_error(y2_test,y2_pred)))
print("R2 Score:",r2_score(y2_test,y2_pred))

print('='*100)
print('Train Hatası (Ezber Kontrolü)')
print('='*100)

y2_pred_train = gbm_reg_cv.predict(X2_train)
print("MAE:",mean_absolute_error(y2_train,y2_pred_train)) 
print("MSE:",mean_squared_error(y2_train,y2_pred_train))
print("RMSE:",np.sqrt(mean_squared_error(y2_train,y2_pred_train)))
print("R2 Score:",r2_score(y2_train,y2_pred_train))
```
