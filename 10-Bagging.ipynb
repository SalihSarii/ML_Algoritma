```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Modeller
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier

#Hata Metrikleri
from sklearn.metrics import accuracy_score,classification_report

#Cross-validation ve en iyi parametrelerin seçimi
from sklearn.model_selection import GridSearchCV,train_test_split

#Kategorik Değişken Dönüşümü
from sklearn.preprocessing import LabelEncoder

# Gereksiz uyarıların görüntülenmemesi için kullanıyoruz
import warnings
warnings.filterwarnings('ignore')

```

```python
df = pd.read_csv('./Data/breast-cancer-wisconsin.csv')
df
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
bag_clf = BaggingClassifier(
    DecisionTreeClassifier(), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1
)
bag_clf.fit(X_train,y_train)
```

```python
y_pred = bag_clf.predict(X_test)
print(classification_report(y_pred, y_test))
```

```python
y_pred_tr = bag_clf.predict(X_train)
print(classification_report(y_pred_tr, y_train))
```

Karar_agaclari gridsearch ile sonuçlar 

train doğruluk oranı : 0.99

test doğruluk oranı : 0.94

```python
bag_clf2 = BaggingClassifier(
    DecisionTreeClassifier(splitter='random',max_leaf_nodes=16), n_estimators=500,
    max_samples=100, bootstrap=True, 
    max_features=0.7, bootstrap_features=True ,n_jobs=-1
)
bag_clf2.fit(X_train,y_train)
```

```python
print('='*100)
print('Test Hatası')
print('='*100)
y_pred = bag_clf2.predict(X_test)
print(classification_report(y_test,y_pred))
print('='*100)
print('Train Hatası (Ezber Kontrolü)')
print('='*100)
y_pred_tr= bag_clf2.predict(X_train)
print(classification_report(y_train,y_pred_tr))
```

```python
rf_clf = RandomForestClassifier(n_estimators=500,max_samples=100,max_leaf_nodes=16,max_features=0.7)
rf_clf.fit(X_train,y_train)
```

```python
print('='*100)
print('Test Hatası')
print('='*100)
y_pred = rf_clf.predict(X_test)
print(classification_report(y_test,y_pred))
print('='*100)
print('Train Hatası (Ezber Kontrolü)')
print('='*100)
y_pred_tr= rf_clf.predict(X_train)
print(classification_report(y_train,y_pred_tr))
```

Bagging Classifier ve RandomForestClassifier neredeyse aynı sonuçları vermektedir. Random Forest'ın Çalışma mantığı bu şekildedir diyebiliriz
