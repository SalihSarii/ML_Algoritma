```python
import pandas as pd
import numpy as np

from sklearn.svm import SVC,SVR

#Train ve Test Setlerinin ayrılması
from sklearn.model_selection import train_test_split

#Değerlendirme Metrikleri
from sklearn.metrics import accuracy_score,classification_report ,mean_absolute_error,mean_squared_error,r2_score

#Cross-validation ve en iyi parametrelerin seçimi
from sklearn.model_selection import GridSearchCV

#Kategorik Değişken Dönüşümü
from sklearn.preprocessing import LabelEncoder

# Gereksiz uyarıların görüntülenmemesi için kullanıyoruz
import warnings
warnings.filterwarnings('ignore')
```

# SVC-  Destek Vektör Sınıflandırma (Support Vector Classification) 

```python
df = pd.read_csv('./Data/breast-cancer-wisconsin.csv')
df
```

```python
df.info()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
```

```python
X = df.drop(columns=['diagnosis'])
y = df['diagnosis']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
svc_clf = SVC()
svc_grid = {
    'kernel' :['linear','poly','rbf'],
    'C' : [100, 10, 1.0, 0.1, 0.001]
}
```

```python
svc_clf_cv = GridSearchCV(svc_clf,param_grid=svc_grid,cv=5,verbose=3)
svc_clf_cv.fit(X_train,y_train)
```

```python
svc_clf_cv.best_params_
```

```python
y_pred = svc_clf_cv.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
y_pred_tr= svc_clf_cv.predict(X_train)
print(classification_report(y_train,y_pred_tr))
```

# SVR-  Destek Vektör Regresyonu (Support Vector Regresyon) 

```python
df2 = pd.read_csv('./Data/insurance.csv')
df2
```

```python
df2.info()
```

```python
df2_encoded = pd.get_dummies(df2,columns=['sex','smoker','region'],dtype=np.int64)
```

```python
X2 = df2_encoded.drop(columns=['charges'])
y2 = df2_encoded['charges']
```

```python
X2_train,X2_test,y2_train,y2_test = train_test_split(X2,y2,train_size=0.8,random_state=42)
```

```python
svr_reg = SVR()
svr_grid = {
    'kernel' :['linear','poly','rbf'],
    'C' : [100, 10, 1.0, 0.1, 0.001]
}
```

```python
svr_reg_cv = GridSearchCV(svr_reg,param_grid=svr_grid,cv=3,verbose=3)
svr_reg_cv.fit(X2_train,y2_train)
```

```python
svr_reg_cv.best_params_
```

```python
y2_pred = svr_reg_cv.predict(X2_test)
print("MAE:",mean_absolute_error(y2_test,y2_pred)) 
print("MSE:",mean_squared_error(y2_test,y2_pred))
print("RMSE:",np.sqrt(mean_squared_error(y2_test,y2_pred)))
print("R2 Score:",r2_score(y2_test,y2_pred))
```

```python
y2_pred_train = svr_reg_cv.predict(X2_train)
print("MAE:",mean_absolute_error(y2_train,y2_pred_train)) 
print("MSE:",mean_squared_error(y2_train,y2_pred_train))
print("RMSE:",np.sqrt(mean_squared_error(y2_train,y2_pred_train)))
print("R2 Score:",r2_score(y2_train,y2_pred_train))
```
